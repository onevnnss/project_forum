## Baixando a base xlsx para o modelo

```python
import pandas as pd
import numpy as np

nome_arquivo = 'receiras_extras/base_de_vendas_2025.xlsx'

try:
    df = pd.read_excel(nome_arquivo, index_col=0) 
    
    df = df.transpose()
    
    print(f"Arquivo '{nome_arquivo}' carregado e ajustado com sucesso!")
    print("Amostra dos dados corrigidos:")
    print(df.head())

except FileNotFoundError:
    print(f"ERRO: O arquivo '{nome_arquivo}' não foi encontrado.")
    print("Por favor, coloque o arquivo na mesma pasta do script e verifique o nome.")
    exit()
print("-" * 80)
```
